```python
# make function to not repeat your work when you deal with tes data
def check_missing_values(df):
        number = df.isnull().sum().sort_values(ascending=False)
        percentage = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
        sns.barplot(number.index[:5],number[:5])
        df_missing = pd.concat([number,percentage],axis=1,keys=['number','percentage'])
        return df_missing
            
    
        
```

```python
## After We remove the most missing colimns now we need to seperater our numirical and catoregical values 
def show_num(df):
    df_num = pd.DataFrame(data=None)
    for label,content in df.items():
        if(pd.api.types.is_numeric_dtype(content)):
            df_num[label] = content
    return df_num

        

    
```

```python
# now do the same for the catogrical values 
def show_string(df):
    df_string= pd.DataFrame(data=None)
    for label,content in df.items():
        if(pd.api.types.is_string_dtype(content)):
            df_string[label] = content
    return df_string
```

## Modeling

```python
x = dftrain.drop('SalePrice',axis=1)
y= dftrain['SalePrice']
np.random.seed(42)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,)
```

```python
def show_scores(model):
    ytprediction = model.predict(x_train)
    yteprediction = model.predict(x_test)
    scores = {
          'Training root mean sequare error':np.sqrt(mean_squared_error(y_train,ytprediction)),
          'Validation root mean sequare error':np.sqrt(mean_squared_error(y_test,yteprediction)),
          'R2 training':r2_score(y_train,ytprediction),
          'R2 Validation':r2_score(y_test,yteprediction),
            'mae Training': mean_absolute_error(y_train,ytprediction),
             'mae valid': mean_absolute_error(y_test,yteprediction)}
    return scores
```

```python
# fit the model 
modelRand = RandomForestRegressor()
modelRand.fit(x_train,y_train)
```

```python
show_scores(modelRand)
```
